```python
import pandas as pd
import jieba
import jieba.analyse
import re
from tqdm import tqdm
tqdm.pandas()
```

```python
train_data =pd.read_csv('data/train_data.csv')
train_data
```

```python
# 删除空行
train_data.dropna(axis=0, how='any', inplace=True)
train_data
```

# 1 删除特殊符号和文本分词

```python

def clearTxt(line):
    if line != '':
        line = line.strip()
        #去除文本中的英文和数字
        line = re.sub("[a-zA-Z0-9]", "", line)
        #去除文本中的中文符号和英文符号
        line = re.sub("[\s+\.\!\/_,$%^*(+\"\'；：“”．]+|[+——！，。？?、~@#￥%……&*（）]+", "", line)
        #分词
        segList = jieba.cut(line, cut_all=False)
        segSentence = ''
        for word in segList:
            if word != '\t':
                segSentence += word + " "
    return segSentence.strip()
train_data['Text'].progress_apply(clearTxt)
```

# 2 去停用词

```python
import codecs
#读取停顿词列表
stopword_list = [k.strip() for k in open(
    'stop/cn_stopwords.txt', encoding='utf8').readlines() if k.strip() != '']
def stopWord(line):
    sentence = delstopword(line, stopword_list)
    return sentence
#删除停用词
def delstopword(line, stopkey):
    wordList = line.split(' ')
    sentence = ''
    for word in wordList:
        word = word.strip()
        if word not in stopkey:
            if word != '\t':
                sentence += word + " "
    return sentence.strip()

train_data['Text'].progress_apply(stopWord)
```

# 3 情感标注的label单独提取

```python
train_data[["Labels"]] = train_data[["Labels"]].astype(int)
train_data[["Labels"]] = train_data[["Labels"]].astype(str)
train_data
```

```python
# 统计最长标签个数
max(list(train_data['Labels'].apply(len)))
```

```python
label_list = []
emo_list = []
train_df = train_data.copy()
train_df
```

```python
# 将情感标注的label全部补充长度统一，并把最后一句话的label提取出来，作为整个数据集的label
for i in range(len(train_data)):
    all_label = str(train_data['Labels'][i])
    single_label = all_label[-1]
    label_list.append(int(single_label))
    # 10表示最长标签个数
    rest_len = 10-len(all_label) 
    if rest_len > 0:
        addzeros = str('0'*rest_len)
        other_label = addzeros+all_label[:-1]
    else:
        other_label = all_label[:-1]
    other_label = '|'.join(list(other_label))
    emo_list.append(str(other_label))
train_df['emo_note'] = emo_list
train_df['label'] = label_list
train_df
```

```python

train_code = train_df.copy()
# 给每个情感标注标签单独作为一列
for i in range(9):
    note_name = 'emo_note_{}'.format(i+1)
    train_code[note_name] = train_code['emo_note'].map(
        lambda x: x.split('|')[i])
train_code = train_code.drop(columns=['ID', 'Labels', 'emo_note'])
train_code
```

```python
train_code.to_csv('data/clear_data.csv',index = False)
```
